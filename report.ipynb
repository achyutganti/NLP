**What difficulties did you run into, if any, while working on this assignment?**<br><br>
This whole assignment can be divided into sub-tasks
* Reading in a text file
* Counting the number of words in that file
* Counting the number of sentences in that file
* Counting the number of syllables in that file
* Plugging the counted values into the formula
* Assess the difficulty with the help of our codified table.

The first three parts of this assignment was easy as we don't have to write heavy logic for that. But the fourth part, counting the syllables, was a bit tricky. The logic was straight but the heuristics that should be considered are many. For example, the rules for counting the syllable said, 
* Count the number of vowels in the word
* Ignore if you have 'e' in the end of that word
* Ignore diphtongs in a word. 

The first case is easy. <br>
Second case seemed easy untill I found a shortcoming. It works fine for words like `forte`, `name` etc. But when you have a word such as `barbeque` it didn't work. Barbeque is a 3 syllable word but according to our logic, we subtract one bacause we have 'e' in the end also we have adjacent vowels `'ue'` in the end. That will subtract one more giving us result = 2. 
So had to deal with that. <br>
Also, words ending with `'le'` had to be handled differently. <br>
Then came words like `'the'`,`'me'`,`'be'` etc. where the only vowel in the word is 'e' and that too in the end. The result will be those words becoming 0 syllable words but they deserve count =1. Had to deal with them separately. <br>
The last two sub-tasks were the easiest of all.
